Firstly we are going to import requests library. Requests allows you to send organic, grass-fed HTTP/1.1 requests, without the need for manual labor.

```
import requests
```

Now we assign the link of the website through which we are going to scrape the data and assign it to variable named website_url.

```
# for example:

url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
website_url = requests.get(url).text
```

requests.get(url).text will ping a website and return you HTML of the website.

We begin by reading the source code for the page and creating a BeautifulSoup (soup)object with the BeautifulSoup function. Beautiful Soup is a Python package for parsing HTML and XML documents. It creates a parse tree for parsed pages that can be used to extract data from HTML, which is useful for web scraping. Prettify() function in BeautifulSoup will enable us to view how the tags are nested in the document.

```
from bs4 import BeautifulSoup
```

```
soup = BeautifulSoup(website_url,'lxml')
print(soup.prettify())
```

If you carefully inspect the HTML script all the table contents i.e. the table in "Toronto - FSAs" which we intend to extract is under class Wikitable Sortable.
![alt text](https://i.pinimg.com/originals/42/67/db/4267db6a8730782c28e22a1f6f460e1a.png)

So our first task is to find class "wikitable sortable"  in the HTML script.

```
my_table = soup.find('table',{'class':'wikitable sortable'})
print(my_table)
```

Remark that Postcode Borough neighborhood are in this format:

< tr>
< td>M1A< /td> <br>
< td>Not assigned< /td> <br>
< td>Not assigned
< /td>< /tr>

I was thinking of saying:

> Before we start spliting those information we should first we should find all links that start with < tr>, we will use find_all('tr')."


but we can write a shorter code. and it is finding all tags that start with td, and loop over evry 3 to extract (Postcode,	Borough neighborhood)

```
links = my_table.findAll('td')
# str(links[5])[4:-6]
links[0]
```

```
f'number of rows in our table is: {len(links)/3}'
```

Now let's loop over them to extract (Postcode, Borough, neighborhood)

```
Postcodes = [] 
Boroughs = [] 
neighborhoods = []

i = 0

for word in links:
  if i%3 == 0:
    Postcodes.append(word.get_text().strip())
    
  elif i%3 == 1:
    Boroughs.append(word.get_text().strip())
  
  elif i%3 == 2:
    neighborhoods.append(word.get_text().strip())
  i+=1
Boroughs
    
```

Let's start creating our data frame regarding those conditions:

> Only process the cells that have an assigned borough. Ignore cells with a borough that is Not assigned


> More than one neighborhood can exist in one postal code area. For example, in the table on the Wikipedia page, you will notice that M5A is listed twice and has two neighborhoods: Harbourfront and Regent Park. These two rows will be combined into one row with the neighborhoods separated with a comma as shown in row 11 in the above table.

>If a cell has a borough but a Not assigned neighborhood, then the neighborhood will be the same as the borough. So for the 9th cell in the table on the Wikipedia page, the value of the Borough and the Neighborhood columns will be Queen's Park.

```
len(neighborhoods) == len(Boroughs) == len(Postcodes)
```

Panda make it easy for us to clean data, so let's dive into pandas dataframe

```
import pandas as pd
```

```
df = pd.DataFrame()
df['Postalcodes'] = Postcodes
df['Borough'] = Boroughs
df['neighborhood'] = neighborhoods
df[df['Postalcodes']== 'M1R']
```

```
indexes = []
for index, row in df.iterrows():
  if row['Borough'] == 'Not assigned':
    indexes.append(index)
    continue
  elif row['neighborhood'] == 'Not assigned':
    row['neighborhood'] == row['Borough']
df.head()
```

```
df.drop(indexes, inplace=True)
df.head()
```

```
df1 = df.groupby('Postalcodes')['neighborhood'].apply(', '.join).reset_index()
df1.head()
```

```
df = pd.merge( df[['Postalcodes','Borough']], df1[['Postalcodes','neighborhood']], on='Postalcodes')
df.head() # you may say: 'emmm why we have two North York!?'=> because we merge on Postalcodes and M34 != M4A
```

Let's explore and cluster the neighborhoods in Toronto.

```
import folium
from geopy.geocoders import Nominatim
```

```
address = 'Toronto'

geolocator = Nominatim(user_agent="t_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Toronto are {}, {}.'.format(latitude, longitude))
```

let's visualizat Toronto neighborhoods

```
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=11)

# add markers to map
for lat, lng, label in zip(df['Latitude'], df['Longitude'], df['neighborhood']):
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```
